# 準備

```python
import pandas as pd
from mplsoccer import Sblocal

pd.set_option("display.max_columns", 100)
```

```python
sb_parser = Sblocal(dataframe=True)
```

# Competition

```python
competition_df = sb_parser.competition(path="../input/open-data/data/competitions.json")

print(competition_df.columns)
print(competition_df.shape)
competition_df.head()
```

```python
competition_df[
    (competition_df["competition_name"]=="FIFA World Cup")&
    (competition_df["competition_gender"]=="male")&
    (competition_df["season_name"]=="2022")
]
```

# Match

```python
match_df = sb_parser.match(path="../input/open-data/data/matches/43/106.json")

print(match_df.columns)
print(match_df.shape)
match_df.head()
```

```python
match_df[
    (match_df["home_team_name"]=="Japan")|
    (match_df["away_team_name"]=="Japan")
]
```

# Event

```python
event_df, related_df, freeze_df, tactics_df = sb_parser.event(path="../input/open-data/data/events/3869219.json")
```

## Event-event

```python
print(event_df.columns)
print(event_df.shape)
event_df.head()
```

```python
pd.crosstab(event_df["type_name"], event_df["team_name"])
```

## Event-related

```python
print(related_df.columns)
print(related_df.shape)
related_df.head()
```

```python
pd.crosstab(related_df["type_name"], related_df["type_name_related"])
```

```python
print(freeze_df.columns)
print(freeze_df.shape)
freeze_df.head()
```

```python
pd.DataFrame(freeze_df.groupby(["player_id", "player_name"])["id"].count()).sort_values(by="id", ascending=False)
```

```python
pd.DataFrame(freeze_df.groupby("id")[["player_id"]].count().value_counts()).sort_index()
```

```python
print(tactics_df.columns)
print(tactics_df.shape)
tactics_df.head()
```

```python
pd.DataFrame(tactics_df.groupby("id")["player_id"].count())
```
